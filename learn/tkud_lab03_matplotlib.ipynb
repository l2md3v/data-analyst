# Matplotlib

`Matplotlib` là một thư viện dùng để vẽ đồ thị cơ bản trong Python, nó cho phép xuất hình ảnh ra các định dạng khác nhau như *jpeg, png*... Trong tài liệu này,chúng ta sẽ tìm hiểu các nội dung:
- Các thành phần cơ bảncủa một đồ thị
- Làm quen với matplotlib
- Các loại đồ thị

## CÁC THÀNH PHẦN CƠ BẢN

Có thể chia một hình vẽ trực quan dữ liệu ra làm 4 lớp như sau:
- Lớp 1: định nghĩa biến số muốn thể hiện (x, y, z...)
- Lớp 2: chọn hình thức thể hiện (đồ thị cột, đồ thị đường, box-plot...)
- Lớp 3: chọn màu sắc và trang trí

Một hình vẽ trên matplotlib được gọi là 1 figure. Cấu trúc một hình vẽ (figure) như sau:

![Anatomy of a figure](https://matplotlib.org/2.0.2/_images/anatomy1.png)

## LÀM QUEN VỚI MATPLOTLIB

Trước hết, để sử dụng, cần phải tải gói thư viện `matplotlib.pyplot`.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### A simple plot

Vẽ đồ thị nối các điểm với giá trị các tung độ lưu trong một list, giá trị hoành độ tương ứng với các chỉ số của list. Chúng ta sử dụng hàm `plot()` như dưới đây

```python
plt.plot([1, 2, 4, 7, 5, 4])
plt.show()
```

Trường hợp này, sử dụng 2 list để lưu các giá trị hoành độ và tung độ của các điểm cần vẽ

```python
plt.plot([-3, -2, 0, 5], [1, 3, 2, 10])
plt.show()
```

Sử dụng hàm `xlim` và `ylim` để  thay đổi giá trị hiển thị đầu và cuối mỗi trục

```python
plt.plot([-3, -2, 0, 5], [1, 3, 2, 10])
plt.xlim(-5, 7)
plt.ylim(None, 12)
plt.show()
```

Tương tự cách sử dụng hàm `plot()` như trên, ta thử vẽ đồ thị hàm $y = x^2$ bằng cách khởi tạo list `x` có 500 giá trị cách đều nhau có giá trị từ -2 đến 2, và list `y` có giá trị là $y = x^2$. Ta vẽ đồ thị  với màu xanh dương (màu sắc mặc định là màu đỏ)

```python
x = np.linspace(-2, 2, 100)
y = x**2

plt.plot(x, y, color='blue')
plt.show()
```

### Title and labels

Sử dụng các hàm `title()`, `xlabel()` và `ylabel()` để thêm tiêu đề cho hình vẽ và tiêu đề cho 2 trục x, y

```python
plt.plot(x, y, color='blue')
plt.title('Hàm bình phương')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)    #Thử thay đổi sang giá trị False để xem kết quả
plt.show()
```

### Figure size

Giả sử bạn muốn vẽ lại đồ thị trên với kích thước khác với kích thước mặc định. Bạn có thể dùng hàm `subplots(figsize=(...))` với tham số là kích thước theo đơn vị inch

Ví dụ: vẽ đồ thị với kích thước 12x4 (inches)

```python
plt.subplots(figsize=(12, 4))

plt.plot(x, y, color='blue')
plt.title('Hàm bình phương')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)    #Thử thay đổi sang giá trị False để xem kết quả
plt.show()
```

### Line style & marker

Thuật ngữ:

- **Marker**: hình dạng thể hiện của các *điểm dữ liệu*
- **Line style**: hình dạng thể hiện của các *đường nối* các điểm dữ liệu với nhau

Mặc định khi vẽ, matplotlib sẽ vẽ bằng cách nối các đường thẳng đến các điểm dữ liệu khác nhau. Tuy nhiên, bạn hoàn toàn có thể thay đổi các giá trị để thay đổi các cách thể hiện này.

```python
plt.plot(x, x*x, 'b--')       # Vẽ đồ thị y=x*x

plt.title('Hàm bình phương')
plt.xlabel('x')
plt.ylabel('y = x**2')

plt.show()
```

Sử dụng đường với các định dạng khác nhau, bạn có thể vẽ nhiều hình trên cùng một đồ thị

Để tham khảo danh sách các định dạng đường (line style) và màu sắc bạn có thể xem ở đây [(LINK)](https://matplotlib.org/2.0.2/api/pyplot_api.html)

Ngoài ra, với mỗi loại định dạng bạn cũng có thể thay đổi độ rộng, độ trong suốt hay định dạng đường nối. Chi tiết bạn có thể tham khảo ở đây [(LINK)](https://matplotlib.org/2.0.2/users/pyplot_tutorial.html)

```python
# Vẽ với các định dạng và màu sắc khác nhau của các linestyle
x = np.linspace(-1.4, 1.4, 30)

plt.plot(x, x, 'g--')   # Đồ thị y = x (g: green)
plt.plot(x, x**2, 'r:') # Đồ thị y = x*x (r: red)
plt.plot(x, x**3, 'b^') # Đồ thị y = x*x*x (b: blue)

plt.show()
```

```python
# Vẽ đồ thị y = x, y = x + 2, y = x + 3, y = x + 4, với độ rộng (lw/ linewidth) khác nhau
plt.subplots(figsize=(6, 4)) 

plt.plot(x, x+1, color='blue', lw=0.25)
plt.plot(x, x+2, color='blue', lw=0.5)
plt.plot(x, x+3, color='blue', lw=1.0)
plt.plot(x, x+4, color='blue', lw=2.0)

plt.title('Plot with different linewidth')
plt.show()
```

```python
# Vẽ đồ thị y = x, y = x + 2, y = x + 3, y = x + 4, với định dạng đường (ls/ linestyle) khác nhau
plt.subplots(figsize=(6, 4)) 

plt.plot(x, x+1, color='blue', ls='-')
plt.plot(x, x+2, color='blue', ls='-.')
plt.plot(x, x+3, color='blue', ls='--')
plt.plot(x, x+4, color='blue', ls=':')

plt.title('Plot with different linestyle')
plt.show()
```

```python
# Vẽ các đồ thị với các điểm đánh dấu (markers) khác nhau
plt.subplots(figsize=(10, 6)) 

plt.plot(x, x+1, color='blue',  ls='-', marker='.')
plt.plot(x, x+2, color='blue',  ls='-', marker=',')
plt.plot(x, x+3, color='blue',  ls='-', marker='o')
plt.plot(x, x+4, color='blue',  ls='-', marker='o', mfc='orange')   # mfc: marker face color
plt.plot(x, x+5, color='blue',  ls='-', marker='o', mec='red')      # mec: marker edge color
plt.plot(x, x+6, color='blue',  ls='-', marker='*', mfc='red')
plt.plot(x, x+7, color='blue',  ls='-', marker='*', mfc='red', markersize=10)
plt.plot(x, x+8, color='blue',  ls='-', marker='+')
plt.plot(x, x+9, color='blue',  ls='-', marker='x')

plt.title('Plot with different markers')
plt.show()
```

### subplots & axes

Trên matplotlib có ba khái niệm là:
- `figure`: là toàn bộ khung cửa sổ của hình vẽ (đồ thị lớn)
- `subplots`: là các đồ thị con bên trong đồ thị lớn (figure). Các subplots được hiển thị theo dạng lưới
- `axes`: là các đồ thị con bên trong đồ thị lớn (figure). Tuy nhiên, các axes có thể được hiển thị theo tọa độ tùy ý người vẽ

Bạn có thể vẽ nhiều đồ thị con trong cùng một đồ thị lớn. Các đồ thị con được tổ chức dạng ma trận bằng các khai báo số dòng, số cột trong hàm `subplot()`. Tham số thứ 3 trong hàm `subplot()` thể hiện chỉ mục của đồ thị được đánh số từ 1 theo thứ tự từ trái sang phải, từ trên xuống dưới.

Chi tiết có thể tham khảo ở đây [(LINK)](https://matplotlib.org/2.0.2/users/gridspec.html)

**Sử dụng subplots**

Subplots in 2x2 grid:

![Subplots in 2x2 grid](https://lectures.scientific-python.org/_images/sphx_glr_plot_subplot-grid_001.png)

Subplots in 1x2 grid:

![Subplots in 1x2 grid](https://lectures.scientific-python.org/_images/sphx_glr_plot_subplot-vertical_001.png)



```python
# Sử dụng suplots vẽ các đồ thị con dạng lưới 2 dòng, 2 cột
x = np.linspace(-1.4, 1.4, 30)      # khởi tạo dữ liệu

plt.subplot(2, 2, 1)
plt.plot(x, x)         # ma trận đồ thị con 2 dòng, 2 cột, đồ thị thứ nhất trái trên

plt.subplot(2, 2, 2)
plt.plot(x, x**2)      # ma trận đồ thị con 2 dòng, 2 cột, đồ thị thứ hai phải trên

plt.subplot(2, 2, 3)   # ma trận đồ thị con 2 dòng, 2 cột, đồ thị thứ ba trái dưới
plt.plot(x, x**3)

plt.subplot(2, 2, 4)
plt.plot(x, x**4)      # ma trận đồ thị con 2 dòng, 2 cột, đồ thị thứ tư phải dưới


plt.show()
```

```python
# Sử dụng suplots vẽ các đồ thị con dạng lưới 2 dòng có 3 đồ thị con
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot = top left
plt.plot(x, x)

plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot = top right
plt.plot(x, x**2)

plt.subplot(2, 1, 2)  # 2 rows, *1* column, 2nd subplot = bottom
plt.plot(x, x**3)
plt.show();
```

Sử dụng **axes**

Subplots in 1x2 grid:

```python
# Sử dụng hàm suplots() vẽ các đồ thị con dạng lưới 2 dòng, 2 cột
x = np.linspace(-1.4, 1.4, 30)      # khởi tạo dữ liệu

fig, ax = plt.subplots(2, 2, constrained_layout=True)

ax[0, 0].plot(x, x, color='orange')         # ma trận đồ thị con 2 dòng, 2 cột, đồ thị thứ nhất trái trên
ax[0, 0].set_title('y = x')

ax[0, 1].plot(x, x**2, color='orange')      # ma trận đồ thị con 2 dòng, 2 cột, đồ thị thứ hai phải trên
ax[0, 1].set_title('$y = x^2$')

ax[1, 0].plot(x, x**3, color='orange')      # ma trận đồ thị con 2 dòng, 2 cột, đồ thị thứ ba trái dưới
ax[1, 0].set_title('$y = x^3$')

ax[1, 1].plot(x, x**4, color='orange')      # ma trận đồ thị con 2 dòng, 2 cột, đồ thị thứ tư phải dưới
ax[1, 1].set_title('$y = x^4$')

fig.suptitle('Vẽ đồ thị con bằng axes object')

plt.show()
```

```python
x = np.linspace(-2, 2, 30)

fig = plt.figure()
fig.suptitle('Vẽ đồ thị con bằng axes object')

ax1 = fig.add_subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot = top left
ax2 = fig.add_subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot = top right
ax3 = fig.add_subplot(2, 1, 2)  # 2 rows, 1 column,  2nd subplot = bottom

ax1.plot(x, x, color='orange')
ax1.set_title('$y = x$')

ax2.plot(x, x**2, color='green')
ax2.set_title('$y = x^2$')

ax3.plot(x, x**3, color='red')
ax3.set_title('$y = x^3$')

plt.show()
```

```python
fig, axs = plt.subplots(ncols=2, nrows=2 # create a suplots in 2x2
                        , figsize=(5.5, 3.5)  # figsize (inches)
                        , layout="constrained"  # giãn cách giữa các axes
                        )

# Hiển thị vị trí các subplot ở chính giữa các đồ thị con
for row in range(2):
  for col in range(2):
    axs[row, col].annotate(f'axs[{row}, {col}]', (0.5, 0.5)
                               ,ha='center', va='center'  # horizotal aligh and vertical align
                               , fontsize=18,
                               color='darkgrey')

fig.suptitle('plt.subplots()')
```

```python
# Vẽ đồ thị trong đồ thị
y = [1, 4, 9, 16, 25,36,49, 64]
x1 = [1, 16, 30, 42,55, 68, 77,88]
x2 = [1,6,12,18,28, 40, 52, 65]

fig = plt.figure()
ax0 = fig.add_axes([0, 0, 1, 1])            # Vẽ subpot có vị trí tại (0, 0) và (width, height) = (1, 1)
ax1 = fig.add_axes([0.1, 0.4, 0.3, 0.3])    # Vẽ subpot có vị trí tại (0.1, 0.4) và (width, height) = (0.3, 0.3)

ax0.plot(x1,y,'bs-') # solid line with yellow colour and square marker
ax1.plot(x2,y,'ro--') # dash line with green colour and circle marker

ax0.set_title("Vẽ đồ thị con bằng axes object")

plt.show()
```

### Chú thích

Bạn có thể thêm bất cứ chữ hay chú thích bằng cách sử dụng hàm `text()`, tham số cần truyền vào là toạ độ x, y và chuỗi cần thêm. Ngoài ra, các tham số khác cũng có thể thay đổi như ví dụ dưới đây như: màu sắc (color), căn lề ngang(horizontalalignment), góc xoay(rotation)...

```python
x = np.linspace(-1.5, 1.5, 30)
px = 0.8
py = px**2

plt.plot(x, x**2, "b-", px, py, "ro")

plt.text(0, 1.5, "Hàm số \n$y = x^2$", fontsize=15, color='blue', horizontalalignment="center")
plt.text(px, py, "x = %0.2f\ny = %0.2f"%(px, py), rotation=10, color='gray')

plt.show()
```

### Nhãn và ghi chú

Bạn có thể thêm nhãn cho mỗi đồ thị vẽ từ hàm `plot()` bằng cách thêm tham số `label`. Sau đó, để hiển thị các nhãn của các đường bạn gọi hàm `legend()` với tham số loc `(location)` chỉ vị trí xuất hiện.

```python
x = np.linspace(-1.4, 1.4, 50)

plt.plot(x, x**2, "r--", label="Square function")
plt.plot(x, x**3, "b-", label="Cube function")

plt.legend(loc="lower right")
plt.show()
```

Một số ví dụ

```python
# creating the dataset
data = {'C':20, 'C++':15, 'Java':30,
        'Python':35}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(courses, values, color ='maroon', width = 0.4)

plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")

plt.show()
```

```python
# Multibar

# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
IT = [12, 30, 1, 8, 22]
ECE = [28, 6, 16, 5, 10]
CSE = [29, 3, 24, 25, 17]

# Set position of bar on X axis
br1 = np.arange(len(IT))                # mảng hoành độ của cột IT
br2 = [x + barWidth for x in br1]       # mảng hoành độ của cột ECE
br3 = [x + barWidth for x in br2]       # mảng hoành độ của cột CSE

# Make the plot
plt.bar(br1, IT, color ='r', width = barWidth,
        edgecolor ='grey', label ='IT')
plt.bar(br2, ECE, color ='g', width = barWidth,
        edgecolor ='grey', label ='ECE')
plt.bar(br3, CSE, color ='b', width = barWidth,
        edgecolor ='grey', label ='CSE')

# Adding Xticks
plt.xlabel('Branch', fontweight ='bold', fontsize = 15)
plt.ylabel('Students passed', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(IT))],
        ['2015', '2016', '2017', '2018', '2019'])

plt.legend()
plt.show()
```

```python
# Stacked bar
N = 5

boys = (20, 35, 30, 35, 27)
girls = (25, 32, 34, 20, 25)
boyStd = (2, 3, 4, 1, 2)
girlStd = (3, 5, 2, 3, 3)
ind = np.arange(N)
width = 0.35

fig = plt.subplots(figsize =(10, 7))
p1 = plt.bar(ind, boys, width, yerr = boyStd)
p2 = plt.bar(ind, girls, width,
             bottom = boys, yerr = girlStd)

plt.ylabel('Contribution')
plt.title('Contribution by the teams')
plt.xticks(ind, ('T1', 'T2', 'T3', 'T4', 'T5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('boys', 'girls'))

plt.show()
```

### Histogram

Vẽ histogram sử dụng hàm `hist()`

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.subplot(1,2,1)
plt.hist(data, bins = 10, rwidth=0.8)

plt.subplot(1,2,2)
plt.hist(data, bins = 4, rwidth=0.5)

plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```

Vẽ histogram với dữ liệu phát sinh từ phân phối chuẩn

```python
x = np.random.normal(170, 10, 1000)     # Phát sinh 1000 số ngẫu nhiên có phân phối chuẩn với mean=170 và std=10

plt.hist(x)

plt.xlabel('Giá trị')
plt.ylabel('Tần số')
plt.title('Histogram')

plt.show()
```

```python
# Thay đổi một số tham số
x = np.random.normal(170, 10, 1000)     # Phát sinh 1000 số ngẫu nhiên có phân phối chuẩn với mean=170 và std=10

plt.hist(x
        , bins=100          # Chia thành 100 cột
        , density=True      # Giá trị tung độ là mật độ
        , stacked=True
        , color='green'     
        , edgecolor='white' # màu cạnh
        )

plt.xlabel('Giá trị')
plt.ylabel('Tần số')
plt.title('Histogram')

plt.show()
```

Vẽ nhiều histogram trên một đồ thị, sử dụng màu khác nhau cùng với giá trị độ trong suốt (alpha)

```python
data1 = np.random.randn(100)
data2 = np.random.randn(100) + 3
data3 = np.random.randn(100) + 6

plt.hist(data1, bins=5, color='g', alpha=0.75, label='bar hist') # default histtype='bar'
plt.hist(data2, color='b', alpha=0.65, histtype='stepfilled', label='stepfilled hist')
plt.hist(data3, color='r', label='bar hist')

plt.legend()
plt.show()
```

Sử dụng hàm `scatter()` để vẽ đồ thị phân tán (scatterplot) của 10 điểm dữ liệu được tạo bởi hàm random như sau:

```python
x, y = np.random.rand(2, 10)
plt.scatter(x, y)
plt.show()
```

Giải thích: hàm `rand(2, 10)`, tạo ra mảng 2 chiều hay nói cách khác là 2 mảng 1 chiều có số cột là 10; giá trị của 2 mảng này được gán cho 2 biến x và y, đại diện toạ độ của 10 cặp điểm.

Bạn có thể thay đổi kích thước bằng cách sử dụng tham số `s` (size) trong hàm `scatter()`

```python
scale = np.random.rand(10)
scale = 500 * scale ** 2
plt.scatter(x, y, s=scale)
plt.show()
```

Hoặc thay đổi màu sắc qua tham số `c`

```python
for color in ['red', 'green', 'blue']:
    n = 10
    x, y = np.random.rand(2, n)
    scale = 500.0 * np.random.rand(n) ** 2
    plt.scatter(x, y, s=scale, c=color, alpha=0.3)

plt.show()
```

### Boxplot

Sử dụng hàm `boxplot()`

```python
data1 = np.random.rand(100)*30      # Phát sinh 100 số ngẫu nhiên trong đoạn [0, 30]

plt.boxplot(x=data1)
plt.title("Boxplot")

plt.show()
```

```python
data1 = np.random.rand(100)*30         # Phát sinh 100 số ngẫu nhiên trong đoạn [0, 30]
data2 = np.random.rand(100)*100        # Phát sinh 100 số ngẫu nhiên trong đoạn [0, 30]

plt.boxplot(x=[data1, data2], labels=['data1', 'data2'])
plt.title("Boxplot")

plt.show()
```
