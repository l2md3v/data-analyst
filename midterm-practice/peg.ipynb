```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
peg = pd.read_csv("penguins-data.csv")
```

```python
peg.head(10)
```

```python
pd.DataFrame(peg["species"].value_counts())
```

```python
pd.DataFrame(peg.groupby("island")["species"].value_counts())
```

```python
peg.groupby("species")[["bill_length_mm", "bill_depth_mm", "flipper_length_mm"]].mean()
```

```python
peg.groupby("species")[
    ["bill_length_mm", "bill_depth_mm", "flipper_length_mm"]
].median()
```

```python
peg.groupby("species")[["bill_length_mm", "bill_depth_mm", "flipper_length_mm"]].std()
```

Nhận xét về kết quả thu được: 
- Chim cánh cụt tập chung nhiều ở đảo Biscoe. 
- Loài Adelie chiếm phần lớn số lượng.

```python
fig, ax = plt.subplots()

for i in peg["species"].unique():
    peg[peg["species"] == i]["island"].hist(ax=ax, alpha=0.5, label=i)


ax.set_xlabel("Island")
ax.set_ylabel("Count")
ax.set_title("Distribution of penguins species on islands")
ax.legend()
plt.show()

print(
    """Nhan xet: 
    - Loai chim Adelie song tren ca 3 dao
    - Loai Gento tap chung o dao Biscoe
    - Loai Chinstrap tap chung o dao Dream"""
)
```

```python
peg.plot(x="bill_length_mm", y="species", kind="scatter")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Species")
plt.title("Penguin Data: Bill Length vs Species")
plt.show()
```

```python
# do thi khao sat moi quan he giua flipper_length va body_mass
peg.plot.scatter(x="flipper_length_mm", y="body_mass_g")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")
plt.title("Flipper Length vs Body Mass")

plt.show()

print(
    """Nhan xet: 
    - Co su tuong quan giua flipper_length va body_mass
    - Kich thuoc flipper_length cang lon thi body_mass cang lon
    """
)
```

```python
fig, ax = plt.subplots()

for i in ["bill_length_mm", "bill_depth_mm"]:
    peg[peg["species"] == "Gentoo"].plot.scatter(
        x=i, y="body_mass_g", ax=ax, alpha=0.5, label=i
    )

ax.set_xlabel("Bill Length/Depth (mm)")
ax.set_ylabel("Body Mass (g)")
ax.set_title("Bill Length/Depth vs Body Mass")
ax.legend()
plt.show()

print(
    """O loai Gentoo, ngoai do dai canh:
    - Co su tuong quan giua bill_length va body_mass
    - Kich thuoc bill_length cang lon thi body_mass cang lon
    """
)
```
